```python
## Imports
import os
import sys
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

data_path = 'D:/Dev/CNN/data/parkinsons'
data_path
```

```python
# Data pathing for spiral
spiral = os.path.join(data_path, 'spiral')
spiral_train = os.path.join(spiral, 'training')
spiral_test = os.path.join(spiral, 'testing')

print(f'Train: {spiral_train} \nTest: {spiral_test}')
```

```python
# Data pathing for wave
wave = os.path.join(data_path, 'wave')
wave_train = os.path.join(wave, 'training')
wave_test = os.path.join(wave, 'testing')
wave_test = wave_test.replace('\\', '/')
print(f'Train: {wave_train} \nTest: {wave_test}')
```

```python
spiral_train_list = [
    os.path.join(spiral_train, classes, image)
    for classes in os.listdir(spiral_train)
    for image in glob.glob(os.path.join(spiral_train, classes, '*'))
    ]
```

```python
# Loading the image
img = mpimg.imread(spiral_train_list[0])
height, width, _ = img.shape

# Display the image using matplotlib.pyplot.imshow
fig, ax = plt.subplots()
img = ax.imshow(img, extent=[0, height, 0, width])
plt.show()
```
